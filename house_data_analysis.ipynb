```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv', encoding='latin1')
df['date'] = df['date'].apply(lambda x: str(x[:8]))
df.head()
```

```python
df.info()
```

```python
df['price_per_sq_lot'] = df['price']/df['sqft_lot']
df.head()
```

```python
df['delta_renovated'] = (df['yr_renovated'] - df['yr_built']).apply(lambda x: 0 if x<0 else x)
df['delta_renovated'] 
```

```python
import calendar
print(calendar.month_name[3])

df['yr_seld']=df['date'].apply(lambda x: int(x[:4]))
df['month_seld']=df['date'].apply(lambda x: calendar.month_name[int(x[4:6])] )
df.head()
```

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace =True)
df.head()
```
