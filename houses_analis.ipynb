Для начала узнаем какие признаки наиболее влияют на стоимость недвижимости, для этого расчитаем матрицу корреляции и построим тепловую диаграмму

```python
plt.figure(figsize=(8, 6))
corr_matrix = np.round(df.corr(), 1)
#Очевидно что наибольше влияют признаки не менее 0.4, их значения мы оставим, а остальные обнулим
corr_matrix[np.abs(corr_matrix < 0.4)] =0
sns.heatmap(corr_matrix, annot=True, linewidth=0.5, cmap='PiYG')
```

```python
plt.figure(figsize=(10, 10))
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg')
plt.title('Price vs sqft living', y=1.0, pad = 0.3)
plt.xlabel('Living area, $m^2$')
plt.ylabel('Price, USD')
plt.grid();
```

Из графика видно что наибольшее количество домов в среднем по стоимости 500000 USD и с жилой площадью 1800 $м^2$ (что отмечалось ранее). Дорогих же домов с площадью более 6000 $м^2$ совсем мало. Наблюдается практически линейная возростающая зависимость цены от жилой площади дома.

```python
plt.figure(figsize=(8, 8))
sns.barplot(x=df['grade'], y=df['price'])
plt.title('Price vs grade')
plt.xlabel('Grade')
plt.ylabel('Price, USD')
plt.grid();
```

Из диаграммы можно сделать вывод что с увеличение качества конструкции и дизайна дома его цена увеличивается по параболе несмотря на небольшие выбросы. Но самые лучшие дома с grade=13 могут иметь очень большую стоимость, что подтверждается выбросом.

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df, x='sqft_above', y='price', palette='deep')
plt.title('Price vs sqft above')
plt.xlabel('Sqft above, $m^2$')
plt.ylabel('Price, USD')
plt.grid();
```

На диаграмме видно очевидную закономерность чем больше площадь дома над землёй по сравнению с подвалом тем он дороже. И зависимость цены дома по сравнению с площадью над землёй практически линейная.

```python
plt.figure(figsize=(10, 10))
sns.jointplot(x=df['sqft_living15'], y=df['price'], kind='kde')
plt.title('Price vs sqft living15', loc='center', y=1.0 , pad = 0.3)
plt.xlabel('Sqft living15 area, $m^2$')
plt.ylabel('Price, USD')
plt.grid();
```

Из диаграммы можно сделать вывод что люди предпочитают, чтобы соседей было поменьше. Но однако наличие соседей влияет на инфроструктуру города и количество учереждений. И поэтому здесь просматривается компромис на вершинах графиков распределения вероятностей sqft_living15 = 1700 $м^2$ и цены 500000 USD. Что говорит о том что люди предпочитают жить в небольших городках с развитой инфроструктурой. Цены таких домов можно и повысить.

```python
plt.figure(figsize=(15, 10))
sns.boxplot(x=df['bathrooms'], y=df['price'], whis=1.5)
plt.title('Price vs bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price, USD')
plt.grid();
```

А здесь с увеличением количества ванных комнат цена увеличивается по пологой параболе. Однако наибольший расброс по цене при количестве ванных 8. Некоторые довольствуются только туалетом. Поэтому дома с bathroom= 0.5 находятся в дешёвой стабильной ценовой категории. Однако оптимальный вариант иметь две ванных комнаты bathroom=2 для четырёх человек в семье. Поэтому в значениях bathrooms в диапазоне от 1.5 до 2.5 наблюдается стабильность в цене.
